```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import torch
```

```python
data= pd.read_csv('/content/water_potability.csv')
data.head()
```

```python
data['ph']=data['ph'].fillna(data['ph'].mean())
data['Sulfate']=data['Sulfate'].fillna(data['Sulfate'].mean())
data['Trihalomethanes']=data['Trihalomethanes'].fillna(data['Trihalomethanes'].mean())
```

```python
numerical = ['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon','Trihalomethanes', 'Turbidity']
scaling = StandardScaler()
scaling.fit(data[numerical])
data[numerical] = scaling.transform(data[numerical])
```

```python
x = data.drop('Potability', axis=1)
y = data['Potability']
```

```python
x = x.to_numpy()
y = y.to_numpy()
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train= torch.tensor(x_train, dtype=torch.float32)
x_test= torch.tensor(x_test, dtype=torch.float32)
y_train= torch.tensor(y_train, dtype=torch.float32)
y_test= torch.tensor(y_test, dtype=torch.float32)
```

```python
x_train.shape[1]
```

####Explain Preprocessing steps

After loading dataset using pandas, first preprocessing step applied missing data handling on three input feature (ph, Sulfate, Trihalomethanes). then,feature scaling was applied using standard scaling on all input features for priotirize all features equally.It's use when there is a large number differences between input features. there is no categorical features in this dataset. So, No categorical encoding required. Then, Separate input features and target features. first convert them to numpy array and then, convert to tensor array for using pytorch library. lastly, we split the test and train data and 20/80 ratio.

```python
import torch.nn as nn
```

```python
class Model(nn.Module):
  def __init__(self,numerical):
   super().__init__()

   self.network= nn.Sequential(
   nn.Linear(numerical,27) ,
   nn.ReLU(),
   nn.Linear(27,18) ,
   nn.ReLU(),
   nn.Linear(18,9) ,
   nn.ReLU(),
   nn.Linear(9,1),
   nn.Sigmoid())

  def forward(self,x):
    out= self.network(x)

    return out
```

## Implement a full training loop:

Forward pass

Backward pass

Gradient reset

### Requirements:

Train for at least 100 epochs.

Store training loss history(You can pick your own Data Structure).

## Write Answer 04:

```python
epochs=500

for epoch in range(epochs):

  y_pred= model(x_train)
  loss= loss_function(y_pred,y_train.reshape(-1,1))
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch % 10 == 0:
    print(f'Epoch: {epoch}, Loss: {loss.item()}')
```

In each steps in the pipeline:



1.   forward propagration was done and got y_pred
2.   then, BCE is calculated by y_pred and y_train which stored in loss variable.

1.  After that, we clear gradient by using zero_grad() before doing back propagation.
2.  New gradient was calculate by using backward() function.

1.   the new weight and biases are updated by using optimizer function.
2.   lastly, we showed losses after each 10 epochs.

*   if learning rate is too small, it takes smaller step to converge and reach to global minima. It need too much time to reach global minima. sometimes,it got stuck on local minima and also It also seems stop on middle for very slow convergence to reach global minima for very small change. Thus, it fail to converge.
