**Question Answer 01**

```python
import numpy as np
import pandas as pd

data=pd.read_csv("/final-employee-ds.csv")
data1=np.array(data["TrainingHours"])
data1=data1.astype(np.float32)
print(data1.mean())
print(data1.std())
```

**Question Answer 02**

```python
data2 = data.copy()
dept_median = data2.groupby("Department")["Salary"].median()
value=data["Department"].map(dept_median)
data2["Salary"]=data2['Salary'].fillna(value)
data2=data2[(data2["ExperienceYears"]>12) & (data2["ProjectCount"]>5)]
print(data2["Salary"].sum())
```

**Question Answer 03**

```python
data3=data.copy()
data3=data3[((data3["Department"]=="IT" )| (data3["Department"]=="Finance"))
             & ((data3["Age"]>=30) & (data3["Age"]<=45))
             & (data3["PerformanceScore"]>88)]
data3.sort_values(by="Salary",ascending=False,inplace=True)
print(data3[["Name","Department","Salary","PerformanceScore"]])
```

**Question Answer 04**

```python
data4=data.copy()
data4["SalaryPerHour"]=data4["Salary"]/(data4["WorkHoursPerWeek"]*4)
data4.sort_values(by="SalaryPerHour",ascending=False,inplace=True)
print(data4[["Name","SalaryPerHour"]].head())
```
