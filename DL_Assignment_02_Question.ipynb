# Question 02: [ Marks 10 ]

What is forward propagation in an artificial neural network? Explain step by step how data moves from the input layer to the output layer.

## **Write Answer 02:**

In Artificial Neural Network, the forward propagation means a process from input to output as forward pass. Here, input data features are taken and going through input layer to output layer without going backwards. Data features given to input layer, then input layer to hidden layers. Lastly, hidden layer to output layer and show the prediction as output.

A neural network has at least 1 Input and 1 output layers.  Hidden layer can be added as needed. Each neuron computes a weighted sum of inputs,
Z = w*x+b (w=weight, x=input, b=bias)

Suppose, for 1 hidden layer, first data features comes to input layer, then from input layer each neuron connect to hidden layer with a weight and bias. Then hidden layer output the weighted sum then apply activation function like relu. Lastly, In output layer, all neuron goes to activation function (Linear, sigmoid, softmax) and last predict an output.

Sigmoid activation function is a function that maps real value to a range 0 to 1. it shows a probability-like output. for 0 input , it gives 0.5 as output. for large positive value , output would be close to 1 and for negative, output would be close to 0. It generally used for binary classification, In ML model like logistic regression, also used for hidden layer also(But now use ReLU).

s(x) = 1 / (1 + e^-x)

*   it can used for binary classification.

**Disadvantages of sigmoid function :**
