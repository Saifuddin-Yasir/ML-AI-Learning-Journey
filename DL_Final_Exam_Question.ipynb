```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset,DataLoader
import kagglehub
import os
```

```python
torch.manual_seed(42)
device = torch.device('cuda'if torch.cuda.is_available() else 'cpu')
print(f" Using Device {device}")
```

```python
data = kagglehub.dataset_download("mohitsingh1804/plantvillage")
```

```python
data
```

```python
torch.manual_seed(42)
device= torch.device('cuda' if torch.cuda.is_available() else 'cpu')

train= os.path.join(data, 'PlantVillage', 'train')
test= os.path.join(data, 'PlantVillage','val')
```

```python
train
```

```python
from torchvision import transforms

transform = transforms.Compose(
    [
        transforms.Resize( (128,128) ),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5]*3, std=[0.5]*3)
    ]
)
```

```python
from PIL import Image

class MultiClassClassfication(Dataset):
  def __init__(self, root_dir,transform=None) :
    super().__init__()

    self.transform=transform
    self.samples =[]


    self.classes = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir,d))])

    self.class_to_idx= {cls_name : idx  for idx,cls_name in enumerate(self.classes)}
    for target_class in self.classes:
      class_index = os.path.join(root_dir, target_class)
      for img_name in os.listdir(class_index):
        path_to_img = os.path.join(class_index, img_name)
        if os.path.isfile(path_to_img):
          label = self.class_to_idx[target_class]
          self.samples.append( (path_to_img,label) )


  def __len__(self):
    return len(self.samples)

  def __getitem__(self,index):
    path_to_img, label = self.samples[index]
    img = Image.open(path_to_img).convert('RGB').convert("RGB")
    if self.transform:
      img = self.transform(img)
    return img,label
```

```python
train_datasetf = MultiClassClassfication(train, transform)
test_datasetf = MultiClassClassfication(test, transform)
num_classes = len(train_datasetf.classes)
print("Number of classes:", num_classes)
print("Full train size:", len(train_datasetf))
print("Full test size:", len(test_datasetf))
```

```python
train = train_datasetf
test= test_datasetf
```

```python
pin = True if device.type == 'cuda' else False
train_loader = DataLoader(train, batch_size=64, shuffle=True, pin_memory=pin)
test_loader = DataLoader(test, batch_size=64, shuffle=False, pin_memory=pin)
```

```python
import torch.nn as nn
import torch.optim as optim

class MyCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.features = nn.Sequential(

            nn.Conv2d(3, 32, kernel_size=4, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(32),
            nn.Conv2d(32, 64, kernel_size=4, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.MaxPool2d(3),
            nn.Conv2d(64, 64, kernel_size=4, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 128, kernel_size=4, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            nn.MaxPool2d(3),
            nn.Conv2d(128, 128, kernel_size=4, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            nn.Conv2d(128, 128, kernel_size=4, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            nn.MaxPool2d(3),
            nn.Conv2d(128, 256, kernel_size=3, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(256),
            nn.Conv2d(256, 256, kernel_size=3, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(256),
            nn.MaxPool2d(3)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, num_classes),
            nn.Softmax()
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x
```

```python
Model= MyCNN(num_classes)
Model.to(device)
lr=0.01
l= nn.CrossEntropyLoss()
optimizer= optim.Adam(Model.parameters(), lr=lr)
epoch= 5
```

```python
for epoch in range(epoch):
  Model.train()
  running_loss= 0
  for batch_features, batch_labels in train_loader:
      batch0_features = batch_features.to(device)
      batch0_labels = batch_labels.to(device)
      optimizer.zero_grad()
      outputs = Model(batch0_features)
      loss_value = l(outputs, batch0_labels)
      loss_value.backward()
      optimizer.step()

      running_loss+= loss_value.item()
  avgloss= (running_loss/len(train_loader))
  print(avgloss)
```

```python
Model.eval()
total, correct = 0, 0
with torch.no_grad():
    for batch_features, batch_labels in test_loader:
        batch_features = batch_features.to(device)
        batch_labels= batch_labels.to(device)
        outputs = Model(batch_features)
        _, predicted = torch.max(outputs, 1)
        total += batch_labels.size(0)
        correct += (predicted == batch_labels).sum().item()

test_acc = correct / total
print(test_acc)
```

# **Explaining each part of network**

- Conv2d layer: it helps to learn features of a image like edges,color, textures.

- ReLU: it's an activation function. which helps to implement non-linearty at learning..which makes model robust.
- BatchNorm2d: It's normalizes used activation function and speed up training. also denied vanishing problem.
- MaxPool2d(3): it's helps to reduces sizes of input layer, also it maintains the data of a range of pixels. it's speed up training withoput losing data which makes model more efficient.
- Flatten: for gives input in ANN part, the data from convulation part 2D convert to 1D.
- Dropout(0.5): 50% neuron drop in each training for save model from overfitting.
- Linear: this layers used in fully connected layer which connected input features to output layers.
- forward(x): it's a user defiend function which connected both classifier and feature blocks. and return output as x.
